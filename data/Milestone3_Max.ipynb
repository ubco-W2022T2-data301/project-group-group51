```python
import pandas as pd
```

```python
df=pd.read_csv("Airline_Data_Jan2021-Dec2021.csv")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bar=sns.barplot(data=df,x='month',y='weather_delay')
```

```python
#This is interesting because I would assume that most weather delays would happen in the winter months. Although February and December
#have slight upticks in weather delays the most delays occur in July and August which is very strange given that these are the months when
#the weather is the best which makes one think that perhaps airlines blame delays on weather when it isn't actually the cause.
```

```python
df['not_carrier_rltd']=df['nas_delay']+df['weather_delay']
df1 = df.groupby('carrier_name')['not_carrier_rltd'].sum().reset_index()
scatter=sns.barplot(data=df1,x='carrier_name',y='not_carrier_rltd')
scatter.set_xticklabels(labels=['Endeavor','American','Alaska','JetBlue','Delta','Frontier','Allegiant','Hawaiian','Envoy','Spirit','PSA','Skywest','Horizon','United','Southwest','Mesa','Republic'],rotation=60)
```

```python
#The most delays not in the control of the carrier comes from Alaska and Southwest airlines. This could be because Alaska especially 
#serves destinations where weather is more extreme then most of the US. And Southwest has a business model that serves smaller,
#less mainstream destinations and airports which typically have worse technology which could result in more NAS delays when technology
#fails.
```

```python
df['carrier_rltd']=df['arr_cancelled']+df['arr_diverted']+df['arr_delay']+df['carrier_delay']+df['security_delay']+df['late_aircraft_delay']
df3=df.groupby('carrier_name')['carrier_rltd'].sum().reset_index()
scatter=sns.barplot(data=df3,x='carrier_name',y='carrier_rltd')
scatter.set_xticklabels(labels=['Endeavor','American','Alaska','JetBlue','Delta','Frontier','Allegiant','Hawaiian','Envoy','Spirit','PSA','Skywest','Horizon','United','Southwest','Mesa','Republic'],rotation=60)
```

```python
#The most carrier related delays happen to Southwest and United. This could be biased because of their large fleet sizes but American 
#and Delta both have larger fleets and incure less delays. A surprising stat is the Allegiant, Frontier, and Spirit have relatively low
#carrier related delays even though they are consistently rated the worst airlines in America. It is not surprising that the regional 
#carriers like envoy, hawaiian and endeavor don't incure many delays since their fleets are small and the smaller aircraft take
#less time to turn around on the ground.
```
