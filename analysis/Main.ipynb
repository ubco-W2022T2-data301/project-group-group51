```python
import pandas as pd
from matplotlib import pyplot as plt

#sample_data = pd.read_csv('airline.csv')
airlineDatajan2022oct2022 = pd.read_csv('Data_files/airlineDatajan2022-oct2022.csv')
#we can changes this file above just ensure naming is somewhat descriptive cause this could get messy
airlineDatajan2022oct2022
```

```python
import numpy as np

flights = airlineDatajan2022oct2022["arr_flights"].sum()
delayed15 = airlineDatajan2022oct2022["arr_del15"].sum()
late = airlineDatajan2022oct2022["late_aircraft_ct"].sum()
cancelled = airlineDatajan2022oct2022["arr_cancelled"].sum()
#diverted = airlineDatajan2022oct2022["arr_diverted"].sum()
#arrivaldelayed = airlineDatajan2022oct2022["arr_delay"].sum()
#carrierdelay = airlineDatajan2022oct2022["carrier_delay"].sum()
#weather = airlineDatajan2022oct2022["weather_delay"].sum()

#, 'diverted', 'arrivaldelayed', 'carrierdelay', 'Weather'

labels = ['Flights', 'Delayed > 15', 'Late', 'Arrival cancelled']
# diverted, arrivaldelayed, carrierdelay, weather
plt.pie([flights, delayed15, late, cancelled], labels = labels, autopct = '%.1f', startangle = 45)
plt.show()
```

```python
plt.bar(airlineDatajan2022oct2022.carrier, airlineDatajan2022oct2022.arr_cancelled)
plt.xlabel('Carrier Code:')
plt.ylabel('Number of flights Canceled')
plt.title('Number Of flights cancelled from Jan 2022 to Oct 2022')
plt.show()
```

```python
plt.bar(airlineDatajan2022oct2022.carrier, airlineDatajan2022oct2022.arr_del15)
plt.title('Number of Flights Delayed 15 Mins or more')
plt.xlabel('Carrier Code:')
plt.ylabel('Number of flights Delayed')
plt.show()
```

```python
plt.bar(airlineDatajan2022oct2022.carrier, airlineDatajan2022oct2022.weather_delay)
plt.xlabel('Carrier Code:')
plt.ylabel('Number of Weather delays')
plt.title('Weather delay bar chart')
plt.show()
```

```python
plt.bar(airlineDatajan2022oct2022.carrier, airlineDatajan2022oct2022.security_delay)
plt.title('Security delays')
plt.xlabel('Carrier Code:')
plt.ylabel('Number of delays')
plt.show()
```

```python
plt.bar(airlineDatajan2022oct2022.carrier, airlineDatajan2022oct2022.late_aircraft_delay)
plt.title('Late aircraft delays')
plt.xlabel('Carrier Code:')
plt.ylabel('Number of late aircraft')
plt.show()
```
