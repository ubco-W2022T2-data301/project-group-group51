```python
import pandas as pd
from matplotlib import pyplot as plt

#sample_data = pd.read_csv('airline.csv')
airlineDatajan2021Dec2021 = pd.read_csv('../data/raw/Airline_Data_Jan2021-Dec2021.csv')
#we can changes this file above just ensure naming is somewhat descriptive cause this could get messy
airlineDatajan2021Dec2021
```

```python
import numpy as np

flights = airlineDatajan2021Dec2021["arr_flights"].sum()
delayed15 = airlineDatajan2021Dec2021["arr_del15"].sum()
late = airlineDatajan2021Dec2021["late_aircraft_ct"].sum()
cancelled = airlineDatajan2021Dec2021["arr_cancelled"].sum()
diverted = airlineDatajan2021Dec2021["arr_diverted"].sum()
arrivaldelayed = airlineDatajan2021Dec2021["arr_delay"].sum()
carrierdelay = airlineDatajan2021Dec2021["carrier_delay"].sum()
weather = airlineDatajan2021Dec2021["weather_delay"].sum()

#, 'diverted', 'arrivaldelayed', 'carrierdelay', 'Weather'

labels = ['Flights', 'Delayed > 15', 'Late', 'Arrival cancelled', 'Diverted', 'Arrival delayed', 'Carrier delay', 'Weather delay']

# diverted, arrivaldelayed, carrierdelay, weather
plt.pie([flights, delayed15, late, cancelled, diverted, arrivaldelayed, carrierdelay, weather], autopct = '%.1f%%', startangle = 90)
plt.legend(labels,title = "Legend",bbox_to_anchor = (1.05,1), loc = "best")
plt.show()
```

```python
plt.bar(airlineDatajan2021Dec2021.carrier, airlineDatajan2021Dec2021.arr_cancelled)
plt.xlabel('Carrier Code:')
plt.ylabel('Number of flights Canceled')
plt.title('Number Of flights cancelled from Jan 2022 to Oct 2022')
plt.show()
```

```python
plt.bar(airlineDatajan2021Dec2021.carrier, airlineDatajan2021Dec2021.arr_del15)
plt.title('Number of Flights Delayed 15 Mins or more')
plt.xlabel('Carrier Code:')
plt.ylabel('Number of flights Delayed')
plt.show()
```

```python
plt.bar(airlineDatajan2021Dec2021.carrier, airlineDatajan2021Dec2021.weather_delay)
plt.xlabel('Carrier Code:')
plt.ylabel('Number of Weather delays')
plt.title('Weather delay bar chart')
plt.show()
```

```python
plt.bar(airlineDatajan2021Dec2021.carrier, airlineDatajan2021Dec2021.security_delay)
plt.title('Security delays')
plt.xlabel('Carrier Code:')
plt.ylabel('Number of delays')
plt.show()
```

```python
plt.bar(airlineDatajan2021Dec2021.carrier, airlineDatajan2021Dec2021.late_aircraft_delay)
plt.title('Late aircraft delays')
plt.xlabel('Carrier Code:')
plt.ylabel('Number of late aircraft')
plt.show()
```
